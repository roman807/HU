## Plot Results

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from pylab import rcParams
import pickle
import glob
import os
os.getcwd()
```

```python
# Load results:
def load_results(name):
    with open(name, 'rb') as f:
        return pickle.load(f)
pkl_data_original = 'results/5ds_1percent/prepared_data.pkl'
data_original = load_results(pkl_data_original)
pkl_relative_freq = glob.glob('results/5ds_1percent/results_relative_freq_*')
assert(len(pkl_relative_freq) == 1)
results_relative_frequency_ = load_results(pkl_relative_freq[0])
pkl_point_difficulty = glob.glob('results/5ds_1percent/results_point_difficulty_*')
assert(len(pkl_point_difficulty) == 1)
results_point_difficulty_ = load_results(pkl_point_difficulty[0])
pkl_semantic_variation = glob.glob('results/5ds_1percent/results_semantic_variation_*')
assert(len(pkl_semantic_variation) == 1)
results_semantic_variation_ = load_results(pkl_semantic_variation[0])
```

```python
### Relative Frequency
rcParams['figure.figsize'] = 8, 5
roc_auc_gbm, roc_auc_iforest, roc_auc_lof = [], [], []
for dataset in ['credit', 'caravan', 'kddcup99']:
    x = np.array(list(results_relative_frequency_[dataset]['gbm'].keys()))
    y_gbm = np.array(list(results_relative_frequency_[dataset]['gbm'].values()))
    y_iforest = np.array(list(results_relative_frequency_[dataset]['iforest'].values()))
    y_lof = np.array(list(results_relative_frequency_[dataset]['lof'].values()))
    roc_auc_gbm.append(y_gbm)
    roc_auc_iforest.append(y_iforest)
    roc_auc_lof.append(y_lof)

mean_roc_auc_gbm = np.mean(np.array(roc_auc_gbm), axis=0)
plt.plot(x, mean_roc_auc_gbm, color='blue', marker='o')
mean_roc_auc_iforest = np.mean(np.array(roc_auc_iforest), axis=0)
plt.plot(x, mean_roc_auc_iforest, color='red', marker='o')
mean_roc_auc_lof = np.mean(np.array(roc_auc_lof), axis=0)
plt.plot(x, mean_roc_auc_lof, color='orange', marker='o')
plt.legend(['GBM','IForest','LOF'])
plt.xlabel('frequency of anomalies')
plt.ylabel('ROC AUC score')
plt.title('ROC AUC Relative frequency')
plt.xticks(x, rotation=45)
plt.show()
```
